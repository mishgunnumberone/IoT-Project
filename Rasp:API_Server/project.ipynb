```python
from flask import Flask
import serial
import time
import json

app = Flask(__name__)

ser = serial.Serial('/dev/ttyACM0', 9600)

def getData(dataType): 
    ser.flushInput()
    data = ser.readline()
    line = data.decode('utf-8')
    line = line[:-1]
    pieces =line.split("\t")
    
    if dataType == "air": 
        temperature = int(pieces[1])
        humidity = int(pieces[0])
        res = {"temperature" : temperature,
           "humidity" : humidity}
        return json.dumps(res)
    elif dataType == "waterLevel": 
        waterLevel = int(pieces[2])
        res = {"waterLevel" : waterLevel}
        return json.dumps(res)
    


@app.route("/air")
def air():
    return getData("air") 

@app.route("/turn_light_on")
def turn_light_on():
    ser.flushInput()
    ser.write("1".encode())
    return ("", 200)

@app.route("/turn_light_off")
def turn_light_off():
    ser.flushInput()
    ser.write("0".encode())
    return ("", 200)

@app.route("/water_level")
def water_level():
    return getData("waterLevel")
    

if __name__ == '__main__':
    app.run(host='0.0.0.0', port='5000')
```
